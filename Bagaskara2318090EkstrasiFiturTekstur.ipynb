# **2. Import Library yang dibutuhkan**

```python
import os
import cv2
import numpy as np
from skimage.feature import graycomatrix, graycoprops
import pandas as pd
```

# **3. Mendefinisikan lokasi folder UTAMA di Google Drive**

```python
  # Folder utama tempat file Colab berada
  folder_proyek = '/content/drive/MyDrive/Komvis/Bagaskara2318090EkstrasiFitur'
  # Folder tempat gambar sampah berada (dalam sub-folder kategori)
  folder_dataset = os.path.join(folder_proyek, 'Garbage Dataset')
  nilai_fitur = []

  print(f"Memulai ekstraksi dari: {folder_dataset}\n")
```

# **4. Melakukan perulangan untuk membaca semua file di folder**

```python
if not os.path.exists(folder_dataset):
    print("⚠️ Folder tidak ditemukan. Pastikan path-nya benar.")
else:
    for file_name in os.listdir(folder_dataset):
        if file_name.lower().endswith((".jpg", ".png", ".jpeg")):

            img_path = os.path.join(folder_dataset, file_name)
            image = cv2.imread(img_path)

            if image is None:
                continue

            # Konversi ke Grayscale
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

            # Ekstraksi fitur GLCM (Jarak 1, Sudut 0, 45, 90, 135)
            glcm = graycomatrix(gray, [1], [0, np.pi/4, np.pi/2, 3*np.pi/4], symmetric=True, normed=True)

            # Menghitung nilai rata-rata
            contrast = graycoprops(glcm, 'contrast').mean()
            dissimilarity = graycoprops(glcm, 'dissimilarity').mean()
            homogeneity = graycoprops(glcm, 'homogeneity').mean()
            energy = graycoprops(glcm, 'energy').mean()
            correlation = graycoprops(glcm, 'correlation').mean()


            # Menampung nilai ke array
            nilai_fitur.append([file_name, contrast, dissimilarity, homogeneity, energy, correlation])
            print(f"✅ Selesai mengekstrak: {file_name}")
```

# **5. Menyusun hasil ke DataFrame dan mengurutkan berdasarkan Nama File**

```python
nama_kolom = ['Nama File', 'Contrast', 'Dissimilarity', 'Homogeneity', 'Energy', 'Correlation']
df = pd.DataFrame(nilai_fitur, columns=nama_kolom)

# Mengurutkan nama file agar mudah diisi manual nanti (IMG 1, IMG 2, dst)
df = df.sort_values(by='Nama File', ignore_index=True)
```

# **6. Menyimpan file Excel di folder utama proyek**

```python
nama_excel = 'Bagaskara2318090_HasilEkstraksi.xlsx'
path_simpan_excel = os.path.join(folder_proyek, nama_excel)
df.to_excel(path_simpan_excel, index=False)

if not os.path.exists(folder_proyek):
    print(f"⚠️ Folder tujuan belum ada. Sistem sedang membuat folder: {folder_proyek}")
    os.makedirs(folder_proyek, exist_ok=True)

# Menyimpan ke Excel
df.to_excel(path_simpan_excel, index=False)

print(f'\n🎉 Proses selesai! File Excel disimpan di: {path_simpan_excel}')
print('👉 Buka file Excel tersebut, dan ketik jenis sampahnya di kolom Kategori!')

# Menampilkan hasil di layar
display(df)
```
